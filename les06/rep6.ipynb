Задание 2.Подготовка данных
Из загруженного набора данных создайте DataFrame, содержащий как признаки, так и целевое значение. Выведите первые 5 строк набора.

```python
diabetes_df = pd.DataFrame(data = diabetes.data, columns = diabetes.feature_names)
diabetes_df['target'] = diabetes.target
diabetes_df.head(5)
```

Задание 4.Исследование данных
Постройте матрицу корреляции.

```python
diabetes_df.corr()
```

Задание 5.Постройте диаграмму рассеяния целевого значение и признака, коэффициент корреляции которого с целевым значением, самый высокий.

```python
max_corr_feature = diabetes_df.corr()['target'].drop('target').idxmax() # Признак с наибольшей корреляцией с целевым значением
sns.scatterplot(x = diabetes_df[max_corr_feature], y = diabetes_df['target'])
plt.show()
```

Задание 6.Сформируйте набор признаков (X) из 5 признаков с самым высоким коэффициентом корреляции с целевым значением. Сформируйте набор для целевого значения (y).

```python
correlation = diabetes_df.corr()['target'] #Вычисл корреляцию с целевым знач
top_5_features = correlation.abs().sort_values(ascending=False).index[1:6]
X = diabetes_df[top_5_features]
y = diabetes_df['target']
print(X, y)
```

Задание 7. Предсказательная модель
Разделите набор данных на два, одни для обучения модели другой для проверки. Тестовый набор должен содержать 25 процентов данных.

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test= train_test_split(X,y,train_size=0.25, random_state=42)
```

Задание 8. Выполните обучение модели.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(y_pred[:5])
print(y_test[:5])
```

Задание 9.Проверка модели
Расчитайте Root mean squared error (RMSE)

```python
from sklearn import metrics
print('Root mean squared error (RMSE) = ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

Задание 10. Расчитайте R² (коэффициент детерминации)

```python
print('R2 = ', np.sqrt(metrics.r2_score(y_test, y_pred)))
```
