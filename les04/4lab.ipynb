5. Выведите распределение переменной Pclass (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса

```python
male = titanik_df['Sex'] == 'male'
female = titanik_df['Sex'] == 'female'

pclass = titanik_df['Pclass']
print("Распредление для всех пассажиров:\n", pclass.value_counts())
print("Распределение для мужчин:\n", pclass[male].value_counts())
print("Распределение для женщин:\n", pclass[female].value_counts())

print("Количество мужчин во втором классе: ", pclass[male].value_counts().loc[2])
```

6. Каковы медиана и среднеквадрати́чное отклонение платежей (Fare)? Округлите до 2 десятичных знаков.

```python
median = round(titanik_df['Fare'].median(), 3)
std = round(titanik_df['Fare'].std(), 3)
print("Медиана ", median)
print("Среднеквадратичное отклонение ", std)
```

7. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
titanik_df_30 = titanik_df[titanik_df['Age'] < 30]
titanik_df_60 = titanik_df[titanik_df['Age'] > 60]
survived30 = round(titanik_df_30['Survived'].mean(), 3)
survived60 = round(titanik_df_60['Survived'].mean(), 3)
print("Процент выживших моложе 30 лет: ", survived30, ", процент выживших старше 60 лет: ", survived60)
print('Правда') if survived30 > survived60 else print('False')
```

8. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

```python
women = titanik_df[titanik_df['Sex'] == "female"]
men = titanik_df[titanik_df['Sex'] == "male"]
survived_women = round(women['Survived'].mean(), 3)
survived_men = round(men['Survived'].mean(), 3)
print("Процент выживших женщин: ", survived_women, ", процент выживших мужчин: ", survived_men)
print('Правда') if survived_women > survived_men else print('False')
```

9. Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
male = titanik_df['Sex'] == 'male'
male_names = titanik_df.loc[male, 'Name'].str.split(" ").str[2]
most_mode_name = male_names.mode().iloc[0]
print(f"{most_mode_name} самое распространненое мужское имя")
```
